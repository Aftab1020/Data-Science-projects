### Popularity Based Recommender System

```python
ratings_with_name = rating.merge(books, on = 'ISBN')
```

```python
ratings_with_name.shape
```

```python
ratings_with_name.head(2)
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns ={'Book-Rating':'num_rating'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns ={'Book-Rating':'avg_rating'},inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_rating','avg_rating','Image-URL-M']]
```

```python
popular_df
```

### Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
readers = x[x].index
```

```python
filter_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(readers)]
```

```python
filter_ratings.head()
```

```python
y = filter_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filter_ratings[filter_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
data = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
data.fillna(0,inplace = True)
```

```python
data
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(data)
```

```python
similarity_score.shape
```

```python
sorted(list(enumerate(similarity_score[0])),key = lambda x:x[1], reverse = True)[1:6]
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(data.index==book_name)[0][0]
    distance = similarity_score[index]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1], reverse = True)[1:6]
    
    for i in similar_items:
        print(data.index[i[0]])
    
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Da Vinci Code')
```
